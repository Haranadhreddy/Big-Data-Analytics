```python
# Print the 9th multiplication table up to 9*9
for i in range(1, 10):
    result = 9 * i
    print(f"9 * {i} = {result}")
```

```python
# Print the multiplication table for 9 as a right-angled triangle
for i in range(1, 10):
    for j in range(1, i + 1):
        result = i * j
        print(f"{i} * {j} = {result}", end="\t")
    print()
```

```python
# Print a 3x5 triangle of *
for i in range(1, 4):
    spaces = " " * (3 - i)  # Add leading spaces to form a triangle
    stars = "*" * (2 * i - 1)  # Number of asterisks in each row
    print(spaces + stars)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbol and the time range
symbol = "^IXIC"
start_date = "2010-01-01"
end_date = "2022-01-01"

# Fetch the stock data from Yahoo Finance
stock_data = yf.download(symbol, start=start_date, end=end_date)

# Extract the adjusted closing prices
closing_prices = stock_data['Adj Close']

# Plotting the graph
plt.figure(figsize=(12, 6))
closing_prices.plot(label='Nasdaq Stock Index')
plt.title('Nasdaq Stock Index (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols and the time range
nasdaq_symbol = "^IXIC"
dow_symbol = "^DJI"
start_date = "2010-01-01"
end_date = "2022-01-01"

# Fetch the stock data from Yahoo Finance for Nasdaq
nasdaq_data = yf.download(nasdaq_symbol, start=start_date, end=end_date)

# Fetch the stock data from Yahoo Finance for Dow Jones
dow_data = yf.download(dow_symbol, start=start_date, end=end_date)

# Extract the adjusted closing prices for both indices
nasdaq_closing_prices = nasdaq_data['Adj Close']
dow_closing_prices = dow_data['Adj Close']

# Plotting the graph
plt.figure(figsize=(12, 6))
nasdaq_closing_prices.plot(label='Nasdaq Stock Index')
dow_closing_prices.plot(label='Dow Jones Industrial Average')
plt.title('Nasdaq and Dow Jones Stock Indices (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols and the time range
symbols = ["GOOGL", "META", "AAPL", "AMZN", "TSLA"]
start_date = "2010-01-01"
end_date = "2022-01-01"

# Fetch the stock data from Yahoo Finance
stock_data = yf.download(symbols, start=start_date, end=end_date)['Adj Close']

# Plotting the graph
plt.figure(figsize=(12, 6))
for symbol in symbols:
    stock_data[symbol].plot(label=symbol)

plt.title('Stock Prices Comparison (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the World Bank API URL for GDP data
api_url = "https://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=excel"

# Fetch the data from the World Bank API
gdp_data = pd.read_excel(api_url, sheet_name="Data", skiprows=3)

# Filter data for the United States and China
us_gdp_data = gdp_data[gdp_data['Country Name'] == 'United States']
china_gdp_data = gdp_data[gdp_data['Country Name'] == 'China']

# Extract the GDP values from 2010 to 2022
us_gdp_values = us_gdp_data.loc[:, '2010':'2022'].values.flatten()
china_gdp_values = china_gdp_data.loc[:, '2010':'2022'].values.flatten()

# Plotting the graph
years = list(map(str, range(2010, 2023)))

plt.figure(figsize=(12, 6))
plt.plot(years, us_gdp_values, label='US GDP', marker='o')
plt.plot(years, china_gdp_values, label='China GDP', marker='o')

plt.title('GDP Comparison: US vs China (2010-2022)')
plt.xlabel('Year')
plt.ylabel('GDP (Current US$)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Monthly expenses data
expenses = {
    'House Mortgage': 2000,
    'Car Lease': 500,
    'Water, Gas, Electricity': 400,
    'Gas': 200,
    'Food': 1100,
    'Leisure': 200,
    'Other Expenses': 600
}

# Calculate the total expense
total_expense = sum(expenses.values())

# Separate the slice for the largest expense (House Mortgage)
explode = [0.1 if expense == 'House Mortgage' else 0 for expense in expenses]

# Plotting the pie chart with actual cost display
plt.figure(figsize=(8, 8))
plt.pie(expenses.values(), labels=expenses.keys(), autopct=lambda p: '{:.0f}'.format(p * total_expense / 100),
        startangle=140, explode=explode)

plt.title('Monthly Expenses Breakdown')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(languages, popularity, color='skyblue')

# Adding text labels above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f'{yval:.1f}', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.ylim(0, max(popularity) + 2)

# Display the bar chart
plt.show()
```
